# Week 13: Exploratory Data Analysis (EDA)

This notebook explores a small study habits dataset using Pandas and Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data/study_habits.csv')
data.head()
```

```python
data.describe()
```

```python
plt.scatter(data['hours_studied'], data['score'])
plt.title('Study Hours vs. Score')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.show()
```

```python
correlations = data[['hours_studied', 'sleep_hours', 'score']].corr()
correlations
```
